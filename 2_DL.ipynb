Calculate the prediction of Revenue on the basis of temperature outside , we have data in Colab Data folder


```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv("/content/drive/MyDrive/Deep learing/Tensorflow/Data/SalesData.csv")
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=df["Temperature"],y=df["Revenue"])
```

```python
x_tarin=df["Temperature"]
y_train=df["Revenue"]
```

```python
import tensorflow as tf
```

```python
model= tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1))
```

```python
model.compile(optimizer= tf.keras.optimizers.Adam(0.5), loss="mean_squared_error")
```

```python
hist_model=model.fit(x_tarin,y_train,epochs=100)
```

```python
hist_model.history.keys()
```

```python
import matplotlib.pyplot as plt
plt.plot(hist_model.history["loss"])
```

AS we can see around 5 iteration there is no need to train it more

```python
model.get_weights()
```

```python
ypred= model.predict([50])
ypred
```

```python
plt.scatter(x_tarin, y_train, color = 'gray')
plt.plot(x_tarin, model.predict(x_tarin), color = 'red')
plt.ylabel('Revenue [dollars]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temperature @Ice Cream Stand')
```
